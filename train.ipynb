```python
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM, InputLayer

physical_devices = tf.config.experimental.list_physical_devices("GPU")
print("Number of GPU: ", len(physical_devices))
print(physical_devices)

if len(physical_devices) != 0:
    tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
x0 = np.load("data/x0.npy")
y0 = np.load("data/y0.npy")
x1 = np.load("data/x1.npy")
y1 = np.load("data/y1.npy")

x0 = x0.reshape(9000, -1, 1)
x0[:10], x0.shape
```

```python
model = Sequential()

model.add(InputLayer(input_shape=(x1.shape[1], 1)))
model.add(LSTM(units=32, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=64, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=64, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=64))
model.add(Dropout(0.2))

model.add(Dense(units=1))

model.compile(optimizer="adam", loss="mean_squared_error")
model.summary()
```

```python
model.fit(x0, y0, epochs=10, batch_size=10)
```

```python
predictions = model.predict(x1)
```

```python
np.save("data/predictions", predictions)
model.save("data/output.h5")
```

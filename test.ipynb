```python
import numpy as np
import tensorflow as tf

physical_devices = tf.config.experimental.list_physical_devices("GPU")
print("Number of GPU: ", len(physical_devices))
print(physical_devices)

if len(physical_devices) != 0:
    tf.config.experimental.set_memory_growth(physical_devices[0], True)


model = tf.keras.models.load_model("data/output.h5")
```

```python
test = np.load("data/x1.npy")[0]
y = np.load("data/y1.npy")

results = []

for i in range(len(y)):
    result = model.predict(test.reshape(-1, 20, 1), verbose=1)
    test = test[1:]
    test = np.append(test, result)
    results.append(result[0])

np.array(results).shape, y.shape
```
